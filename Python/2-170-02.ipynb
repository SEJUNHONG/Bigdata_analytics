```python
data=pd.read_csv("Hitters.csv", index_col = 0)
data.head()
```

```python
data.info()
```

```python
med=data['Salary'].median()
med
```

```python
data['Salary'].describe()
```

```python
IQR=750-190
IQR
```

```python
upper_bound=med+2*IQR
lower_bound=med-2*IQR
```

```python
data=data[data['Salary'].notnull()]
```

```python
data=data[(data['Salary']>upper_bound) | (data['Salary']<lower_bound)]
data
```
