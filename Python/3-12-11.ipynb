```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Boston_housing.csv", index_col=0)
data.head()
```

```python
data.info()
```

```python
data1=data.sort_values(by='crim', ascending=False)[:10]
top10=data1['crim']
top10
```

```python
top_crim10=top10.iloc[9]
top_crim10
```

```python
data['crim']=np.where(data['crim']>=top_crim10, top_crim10, data['crim'])
data
```

```python
data['crim'].value_counts()
```

```python
data_age80=data[data['age']>=80]
data_age80
```
