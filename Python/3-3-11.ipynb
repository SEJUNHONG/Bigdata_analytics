```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Boston_housing.csv", index_col=0)
data.head()
```

```python
data.info()
```

```python
min_medv=np.min(data['medv'])
min_medv
```

```python
data2=data.sort_values(by='medv', ascending=False)
data2
```

```python
medv_top50=data2['medv'][50]
medv_top50
```

```python
data['medv']=np.where(data['medv']>medv_top50, min_medv, data['medv'])
```
