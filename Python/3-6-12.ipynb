```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Cars93.csv", index_col=0)
data.head()
```

```python
data.info()
```

```python
data2=data[data['Luggage.room'].notnull()]
data2
```

```python
med=np.median(data2['Luggage.room'])
med
```

```python
before=np.mean(data2['Luggage.room'])
before
```

```python
data['Luggage.room']=np.where(data['Luggage.room'].isnull(), med, data['Luggage.room'])
```

```python
after=np.mean(data['Luggage.room'])
after
```

```python
print(after-before)
```
