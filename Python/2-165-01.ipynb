```python
data=pd.read_csv("Boston_housing.csv", index_col = 0)
data
```

```python
top10=data.sort_values(by='crim' ,ascending=False)[:10]
top10
```

```python
tenth=min(top10['crim'])
tenth
```

```python
data['crim']=np.where(data['crim']>=tenth, tenth, data['crim'])
data['crim']
```
