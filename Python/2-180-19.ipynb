```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("ChickWeight.csv", index_col=0)
data.head()
```

```python
data.info()
```

```python
train=data[data['Time']==10]
train
```

```python
before=np.mean(train['weight'])
before
```

```python
train2=train.sort_values(by='weight', ascending=False)[:30]
train2
```

```python
new=np.min(train2['weight'])
new
```

```python
train['weight']=np.where(train['weight']>new, before, train['weight'])
train
```

```python
after=np.mean(train['weight'])
after
```

```python
print(before-after)
```
