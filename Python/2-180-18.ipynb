```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("marvel-wikia-data.csv", index_col=0)
data.head()
```

```python
data.info()
```

```python
data['HAIR'].value_counts()
```

```python
data['EYE'].value_counts()
```

```python
train=data[(data['HAIR']=='Brown Hair') & (data['EYE']=='Brown Eyes')]
train
```

```python
mean=np.mean(train['APPEARANCES'])
mean
```

```python
std=np.std(train['APPEARANCES'])
std
```

```python
lower_bound=mean-1.5*std
lower_bound
```

```python
upper_bound=mean+1.5*std
upper_bound
```

```python
train_filter=train[(train['APPEARANCES']>lower_bound) & (train['APPEARANCES']<upper_bound)]
train_filter
```

```python
print(np.mean(train_filter['APPEARANCES']))
```
