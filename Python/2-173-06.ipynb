```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
data.head()
```

```python
data.info()
```

```python
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data['TotalCharges'].isnull().sum()
```

```python
data['TotalCharges']=np.where(data['TotalCharges'].isnull(), 0, data['TotalCharges'])
```

data=data.astype({'TotalCharges':'float64'})
data.info()

```python
data['TotalCharges'].describe()
```

```python
q1=data['TotalCharges'].describe().loc['25%']
q3=data['TotalCharges'].describe().loc['75%']
```

```python
mean=q1=data['TotalCharges'].describe().loc['mean']
```

```python
data=data[(data['TotalCharges']<(mean+1.5*q3)) & (data['TotalCharges']>(mean-1.5*q1))]
data
```
